# Build systems

## Compiling "Hello world!"

```python
%%bash
cat >hello.c <<EOF
#include <stdlib.h>
#include <stdio.h>

int main() {
  printf("Hello world!");
  return EXIT_SUCCESS;
}
EOF
```

An executable file `hello` can be built from the source file `hello.c` by a single call to `gcc`.

```python
%%bash
gcc -o hello hello.c
./hello
```

## GCC as a compilation chain

The `gcc` command automatically invoquates the sequence of transformation tools that transform the source code into an executable file: compilation, assembling, linking...

```python
%%bash
gcc --verbose -o hello hello.c
```

These compilation steps can be made explicit: that can be useful to track the changes and understand the compiler behaviour.

- Preprocessing (expand macros and #... directives):

```python
%%bash
gcc -E -o hello.pp.c hello.c
cat hello.pp.c
```

- Compilation:

```python
%%bash
gcc -S -o hello.s hello.pp.c
cat hello.s
```

- Assembling:

```python
%%bash
gcc -c -o hello.o hello.s
file hello.o
```

- Linking:

```python
%%bash
gcc -o hello hello.o
file hello
```

## Adding dependencies

Compilation often involves some sophistifications.  For instance, during tutorials, our program will rely on the GTK library.  Here is an example reduced to the invokation to the library initializer.

```python
%%bash
cat >gtk_example.c <<EOF
#include <stdlib.h>
#include <gtk/gtk.h>

int main() {
  gtk_init(0, NULL);
  return EXIT_SUCCESS;
}
EOF
```

The compilation of this example fails as the preprocessor does not know where to find the file `gtk/gtk.h`.

```python
%%bash
gcc -o gtk_example.o gtk_example.c
```

The compilation phase succeeds if we enumerate the collection of include paths that covers the GTK library and its dependencies.

As such an enumeration is cumbersome and is specific to a given installation, we may use the tool `pkg-config` that enumerates the include paths relative to a given library for us.

```python
%%bash
pkg-config --cflags gtk+-2.0
```

```python
%%bash
gcc -c -o gtk_example.o gtk_example.c `pkg-config --cflags gtk+-2.0`
```

The linking phase also fails as the symbols that the library defined are visible only if the library is linked with our program.

```python
%%bash
gcc -o gtk_example gtk_example.o
```

`pkg-config` can also be asked to list the libraries to be linked with.

```python
%%bash
pkg-config --libs gtk+-2.0
```

```python
%%bash
gcc -o gtk_example gtk_example.o `pkg-config --libs gtk+-2.0`
```

## Building chain

![Building chain](build/compilation_chain.png)

Such a chain is a particular case of a dependency graph. A dependency graph is a directed acyclic graph that explains how to build the project targets (the executable, the end-user form of the documentation, the test reports, etc.) from the source files.  Directed because the arrows transform a source into a result.  Acyclic because the compilation process should terminate.

**Note.** There exist some specific cases where some cycles naturally appear in a compilation process.  These cases are known as “bootstrapping”, where a project develops a tool that is used to build the tool itself.  Typically, C compilers are now usually written in C (or C++), and we need a C compiler to compile them (and a compiler developper often wants to compile her compiler by using her own compiler).  In that cases, the cyclicity should be broken by adding a “bootstrap” step that provides a seed (for example, a compiler already compiled) to initiate the building process.  There will then appear a meta-graph of dependencies between builds, where the arrows that involve the tool for their transformations will rely on the results of the previous build.

## Building graph

```python
%%bash
cat >gtk_example.c <<EOF
#include <stdlib.h>
#include "initialization.h"

int main() {
  initialization();
  return EXIT_SUCCESS;
}
EOF
```

```python
%%bash
cat >initialization.c <<EOF
#include "initialization.h"
#include <gtk/gtk.h>

void initialization(void) {
  gtk_init(0, NULL);
}
EOF
```

```python
%%bash
cat >initialization.h <<EOF
#ifndef _initialization_h_
#define _initialization_h_

void initialization(void);

#endif /* _initialization_h_ */
EOF
```

![Compilation graph](build/compilation_graph.png)

- Preprocessing, Compilation, Assembling

```python
%%bash
gcc -c -o gtk_example.o gtk_example.c
gcc -c -o initialization.o initialization.c `pkg-config --cflags gtk+-2.0`
```

- Linking

```python
%%bash
gcc -o gtk_example gtk_example.o initialization.o `pkg-config --libs gtk+-2.0`
```

## Makefiles: defining rules for build arrows

```python
%%bash
cat >Makefile <<EOF
all: gtk_example

clean:
	rm -f gtk_example gtk_example.o

gtk_example.o: gtk_example.c initialization.h
	gcc -c -o gtk_example.o gtk_example.c `pkg-config --cflags gtk+-2.0`

initialization.o: initialization.c initialization.h
	gcc -c -o gtk_example.o gtk_example.c `pkg-config --cflags gtk+-2.0`

gtk_example: gtk_example.o initialization.o
	gcc -o gtk_example gtk_example.o initialization.o `pkg-config --libs gtk+-2.0`
EOF
```

```python
%%bash
make
```

Makefiles are architecture specific: as rules are defined as shell commands, they are specific to a POSIX environment.  Higher level tools allow us to define buuilding graphs more declaratively and more portably.  Historically, autotools are such higher level tools to add some declarativity for building graphs in the POSIX world.  CMake is a more modern tool that can generate, from a single project definition, either Unix Makefiles, or Visual C++ projects for Windows, or XCode projects for Mac OS X, etc.

## Minimal usage of CMake

A `CMakeLists.txt` is a list of simple commands that declare which are the source files and which dependencies the project relies on.

```python
%%bash
cat >CMakeLists.txt <<EOF
project(gtk_example)
cmake_minimum_required(VERSION 2.8)
cmake_policy(VERSION 2.8)

set(SOURCES gtk_example.c initialization.c)

add_executable(gtk_example \${SOURCES})

find_package(GTK2 2.6 REQUIRED)
include_directories(\${GTK2_INCLUDE_DIRS})
target_link_libraries(gtk_example \${GTK2_LIBRARIES})
EOF
```

To build a project which building has been defined with CMake in a Unix environment, it is usual to create a fresh `build/` directory to work in.  The `cmake` tool takes the path of the project (where the `CMakeLists.txt` lives in) as parameter and generates a `Makefile` according to the current architecture.

```python
%%bash
mkdir -p build/
cd build/
cmake ..
```

```python
%%bash
cd build/
make
```

- The command `project()` gives a name to the project.
- The commands `cmake_minimum_required()` and `cmake_policy()` declare the version of CMake for which the `CMakeLists.txt` has been written to ensure ascending compatibility.
- The command `set()` defines a variable.  Note that the name `SOURCES` is not magic: it just gives a name to the list of sources files that we will be able to refer as `${SOURCES}` in the following lines.  It is worth noticing that the order of commands in the list is relevant: the variable will not be defined (i.e., will be empty) before its definition.
- The command `add_executable()` adds a compilation target to be built: the first argument is the name of the executable and the other arguments list the source files.  C is the default compiler.
- The command `find_package()` tells CMake to import the definitions from CMake packages that typically declare variables relevant for the use of external libraries.  Note that to ensure ascending compatibility and robustness, the version is specified.  The `REQUIRED` argument asks CMake to fail if the library is not found in the current system where the project is built.  Some packages can be made `OPTIONAL` if the project can be built without them.
- The commands `include_directories()` and `target_link_libraries()` add lists of include directories or libraries for the compiler or the linker respectively.  Note that `target_link_libraries()` takes as argument the name of the target (i.e., the executable) which the libraries have to be linked to.

## Using CMake to build a library

It is often useful to create our own libraries to gather some compilation units together, either for our own project or to distribute our library to other developers.  The `add_library` commands allow to define a library as compilation target, either to be linked statically with the executable (`STATIC`), or dynamically (`SHARED`).

```python
%%bash
cat >CMakeLists.txt <<EOF
project(gtk_example)
cmake_minimum_required(VERSION 2.8)
cmake_policy(VERSION 2.8)

add_library(gtk_example_lib STATIC initialization.c)
add_executable(gtk_example gtk_example.c)

target_link_libraries(gtk_example gtk_example_lib)

find_package(GTK2 2.6 REQUIRED)
include_directories(\${GTK2_INCLUDE_DIRS})
target_link_libraries(gtk_example_lib \${GTK2_LIBRARIES})
EOF
```

Note that we declared `gtk_example_lib` in the list of libraries that `gtk_example` has to be linked to.  On the other hand, `${GTK2_LIBRARIES}` is linked to `gtk_example_lib`.

# Annex: additional features for Makefiles

```python
%%bash

cat >Makefile <<EOF
SOURCES=gtk_example.c initialization.c
CFLAGS=\$(shell pkg-config --cflags gtk+-2.0)
LDLIBS=\$(shell pkg-config --libs gtk+-2.0)
    
all: gtk_example
    
clean:
	rm -f gtk_example \$(SOURCES:.c=.o)
    
gtk_example: \$(SOURCES:.c=.o)
EOF
```

```python
%%bash
make
```

In a `Makefile`, the pseudo-variable `$(shell `some commands`)` is expanded to the output of that shell command: the variables `$(CFLAGS)` and `$(LDLIBS)` are therefore bound to the arguments provided by `pkg-config` to compile and link the sources properly with the GTK library.

In a `Makefile`, the expansion of `$(`variable`:`old suffix`=`new suffix`)` replaces in every word of `$(`variable`)` the old suffix by the new suffix: therefore, `$(SOURCES:.c=.o)` will be expanded into `tetris.o`, `grid.o`, `window.o`.

`make` defines implicit rules to compile and link C programs.

    %.o: %.c
    	$(CC) -o $@ -c $(CPPFLAGS) $(CFLAGS) $<

`%` introduces patterns for generic rules, `$<` refers to the input files of the rule, `$@` to the target file.

And single objects like `tetris` are implicitly built with `$(CC) $(LDFLAGS) -o $@ $< $(LOADLIBS) $(LDLIBS)`.
